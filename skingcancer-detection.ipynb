```python
#SKIN CANCER DETECTION 
# IMPORT libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
Breast_Cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
# if you want to drop a column  do axis = 1
x = data_Frame.drop(columns ='label',axis = 1)
y = data_Frame['label']
```

```python
X_train ,X_test ,Y_train ,Y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
print(x.shape,X_train.shape,X_test.shape)
```

```python
#standarise the data 
from sklearn.preprocessing  import StandardScaler
scaler = StandardScaler()
x_train_std = scaler.fit_transform(X_train)
x_test_std = scaler.transform(X_test)
```

```python
x_train_std
```

```python
# Creating a simple neural network

# Importing TensorFlow and Keras
import tensorflow as tf
from tensorflow import keras

# Setting a seed for reproducibility
# This ensures consistent results across multiple runs
tf.random.set_seed(3)

# Note: Setting the seed helps avoid variation in accuracy due to random factors
# during training, such as weight initialization.
```

```python
#seting up the layers of my neural network 
model = keras.Sequential([
                          keras.layers.Flatten(input_shape =(30,)),
                          keras.layers.Dense(20,activation = 'relu'),
                          keras.layers.Dense(2,activation='sigmoid')
                         ])
```

```python
#Neural network 
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

```

```python
# Training 
history = model.fit(x_train_std, Y_train, validation_split = 0.2, epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model acuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model acuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','validation data'], loc = 'upper right')
```

```python
# Acuuracy of the model on the test data 
loss , accuracy = model.evaluate(x_test_std, Y_test)
print(accuracy)
```
